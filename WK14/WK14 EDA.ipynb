# Body Mass Index (BMI)

```python
df['Height'] = df['Height'] / 100
df['BMI'] = df['Weight'] / (df['Height'] ** 2)

def categorize_bmi(bmi):
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 25:
        return "Normal weight"
    elif 25 <= bmi < 30:
        return "Overweight"
    else:
        return "Obese"
    
df['BMI Category'] = df['BMI'].apply(categorize_bmi)
bmi_counts = df['BMI Category'].value_counts()

plt.figure(figsize=(8, 8))
plt.pie(
    bmi_counts, 
    labels=bmi_counts.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=['lightblue', 'lightgreen', 'orange', 'red']
)
plt.title("Overall BMI Distribution")
plt.show()
```

Split Into Genders

```python
df_male = df[df['Gender'] == 'Male']
df_female = df[df['Gender'] == 'Female']

bmi_counts_male = df_male['BMI Category'].value_counts()
bmi_counts_female = df_female['BMI Category'].value_counts()

fig, axes = plt.subplots(1, 2, figsize=(14, 7))

axes[0].pie(
    bmi_counts_male, 
    labels=bmi_counts_male.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=['lightblue', 'lightgreen', 'orange', 'red']
)
axes[0].set_title('BMI Distribution for Males', fontweight = "bold")

axes[1].pie(
    bmi_counts_female, 
    labels=bmi_counts_female.index, 
    autopct='%1.1f%%', 
    startangle=140, 
    colors=['lightblue', 'lightgreen', 'orange', 'red']
)
axes[1].set_title('BMI Distribution for Females', fontweight = "bold")

plt.tight_layout()
plt.show()
```

Activity Level by Gender

```python
genders = sorted(df["Gender"].unique())

fig, axes = plt.subplots(1, 2, figsize=(12, 5), sharey=True)

colors = {
    "Lightly Active": "blue",
    "Sedentary": "red",
    "Moderately Active": "green",
    "Very Active": "orange",
    "Extremely Active": "purple"
}

for i, gender in enumerate(genders):
    subset = df[df["Gender"] == gender]["Activity Level"].value_counts(normalize=True) * 100

    subset = subset.reindex(colors.keys(), fill_value=0)

    axes[i].bar(subset.index, subset.values, color=[colors[act] for act in subset.index])
    axes[i].set_title(f"{gender}", fontsize=12, fontweight="bold")
    axes[i].set_xticklabels(subset.index, rotation=20)
    axes[i].set_ylim(0, 100)

fig.suptitle("Activity Level Distribution by Gender", fontsize=14, fontweight="bold")

axes[0].set_ylabel("Percentage (%)")

plt.tight_layout()
plt.show()
```

Daily Calorie Target by Activity Level

```python
color = ['#6D6875','#B5838D','#E5989B','#FFB4A2','#FFCDB2']
activity_levels = df['Activity Level'].unique()
avg_calories = [df[df['Activity Level'] == level]['Daily Calorie Target'].mean() for level in activity_levels]

sorted_indices = np.argsort(avg_calories)
activity_levels = np.array(activity_levels)[sorted_indices]
avg_calories = np.array(avg_calories)[sorted_indices]

plt.figure(figsize=(9, 6))
plt.barh(activity_levels, avg_calories, color=color)

plt.xlabel('Average Daily Calorie Target', fontweight = "bold")
plt.ylabel('Activity Level', fontweight = "bold")
plt.title('Average Daily Calorie Target by Activity Level', fontweight = "bold")

plt.grid(axis='x', linestyle='-', alpha=0.7)
plt.show()
```

Daily Calorie Target by Dietary Preference

```python
# Step 1: Calculate required protein intake
df['Required Protein Intake'] = df['Weight'] * 2.2

# Step 2: Create a column to check if they meet their requirement
df['Meets Protein Target'] = df['Protein'] >= df['Required Protein Intake']

# Step 3: Count how many meet vs. don't meet the target
counts = df['Meets Protein Target'].value_counts()

# Step 4: Plot the pie chart
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=['Meets Target', 'Does Not Meet Target'], autopct='%1.1f%%', colors=['#98FF98', '#FFB4A2'])
plt.title('Percentage of People Who Meet Their Protein Intake', fontweight = "bold")
plt.show()
```

Percentage of who is maintaing weight, losing weight, and gaining weight

```python
df['Estimated Maintenance Calories'] = df['Weight'] * 30.8
calorie_diff = df['Daily Calorie Target'] - df['Estimated Maintenance Calories']

weight_status = []
for diff in calorie_diff:
    if diff > 200:
        weight_status.append('Gaining Weight')
    elif diff < -200:
        weight_status.append('Losing Weight')
    else:
        weight_status.append('Maintaining Weight')

df['Weight Status'] = weight_status

counts = df['Weight Status'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['#dd6e42', '#e8dab2', '#4f6d7a'])
plt.title('Weight Status Based on Calorie Intake', fontweight = "bold")
plt.show()
```
